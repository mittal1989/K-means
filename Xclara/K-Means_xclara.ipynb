```python
#Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importing the dataset
data = pd.read_csv('C:/Users/lokesh.saini/Downloads/friendly-fortnight-master/xclara.csv')
x = data.iloc[:,[0,1]].values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range (1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300,n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Curve')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#Fitting data to kmeans
kmeans = KMeans(n_clusters = 3, init = 'k-means++',max_iter = 300,n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

```python
#Visualising the CLusters
plt.scatter(x[y_kmeans == 0,0], x[y_kmeans == 0,1], s = 100, c = 'red',edgecolors = 'black', label = 'Cluster 1')
plt.scatter(x[y_kmeans == 1,0], x[y_kmeans == 1,1], s = 100, c = 'green',edgecolors = 'black', label = 'Cluster 2')
plt.scatter(x[y_kmeans == 2,0], x[y_kmeans == 2,1], s = 100, c = 'blue', edgecolors = 'black', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow',edgecolors = 'black', label = 'Centroids')
plt.title('Clusters')
plt.xlabel('V1')
plt.ylabel('V2')
plt.legend()
plt.show()
```
